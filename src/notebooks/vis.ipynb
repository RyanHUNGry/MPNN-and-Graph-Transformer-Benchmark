We experiment with hidden layer count on GAT and GCN for Cora in `run.py`. Here is the visualization of results.

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Data
hidden_layers = [2, 4, 6, 8]
gcn_accuracies = [(1.0, 0.796), (0.7143, 0.651), (0.6429, 0.585), (0.5786, 0.408)]
gat_accuracies = [(0.779, 0.781), (0.781, 0.767), (0.767, 0.751), (0.751, 0.738)]

# Split train and test accuracies
gcn_train, gcn_test = zip(*gcn_accuracies)
gat_train, gat_test = zip(*gat_accuracies)

# Plot
plt.figure(figsize=(10, 6))
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# GCN group with distinct colors and markers
plt.plot(hidden_layers, gcn_train, label='GCN Train Accuracy', marker='o', linestyle='solid', color='blue')
plt.plot(hidden_layers, gcn_test, label='GCN Test Accuracy', marker='o', linestyle='solid', color='purple')
# GAT group with distinct colors and markers
plt.plot(hidden_layers, gat_train, label='GAT Train Accuracy', marker='s', linestyle='dashdot', color='red')
plt.plot(hidden_layers, gat_test, label='GAT Test Accuracy', marker='s', linestyle='dashdot', color='orange')

# Labels and legend
plt.title('Train/Test Accuracies for GCN and GAT')
plt.xlabel('Number of Hidden Layers')
plt.ylabel('Accuracy')
plt.xticks(hidden_layers)
plt.ylim(0, 1)  # Set y-axis limits for better visualization
plt.legend()
plt.grid(True)
plt.show()
```
